```python
import pandas as pd
```

```python
df = pd.read_csv('../data/dataset.csv')
df.head(10)
```

```python
df = df.drop(columns=['Unnamed: 11','Furnished_status'])
df.head(10)
```

```python
df['Balcony']=df['Balcony'].fillna(0)
df['parking']=df['parking'].fillna(0)
df['Lift']=df['Lift'].fillna(0)
df['type_of_building']=df['type_of_building'].fillna('Flat')
df.head(5)
```

```python
dummies = pd.get_dummies(df[['neworold','type_of_building']],).drop(columns=['neworold_Resale','type_of_building_Individual House'])
dummies.head(5)
```

```python
newdf= df.drop(columns=['neworold','type_of_building','Price_sqft'])
final  = pd.concat([newdf,dummies],axis='columns')
final.head(5)
```

```python
x=final.drop('price',axis='columns')
y=final['price']
```

```python
from sklearn.linear_model import LinearRegression
import joblib
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
joblib.dump(model,'price_prediction_model.joblib')
```

```python
round(model.predict([[1000,2,1,2,1,1,True,True]])[0],2)
```
